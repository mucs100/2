```python
import pandas as pd
df=pd.read_csv('/homeprices_multiple.csv') #Reading csv file
df
```

```python
#Finding median of BEDROOMS column
df.BEDROOMS.median()
```

```python
#Replacing NaN value with Median value
df.BEDROOMS=df.BEDROOMS.fillna(df.BEDROOMS.median())
#Displaying dataset after replacing NaN value
df
```

```python
x=df.iloc[:,[0,2,3]].values #selecting features
y=df.iloc[:,1].values #selecting label
```

```python
#Creating  model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
```

```python
#Predicting the value
model.predict([[3200,18,610000]])
```

```python
#finding coefficient
model.coef_
```

```python
#finding intercept
model.intercept_
```

```python
#finding slope
4.03512705e-04*3200+(-7.52257093e-02)*18+3.69935422e-06*610000+1.6308078063185052
```
